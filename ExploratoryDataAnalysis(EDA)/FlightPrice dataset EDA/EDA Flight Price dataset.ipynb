#now we will handle the datatype problem : they re all object we need to correct it 
:as we are doing eda of this data set so that our ml model can understand the data correctly ; 
it only understands the numerical values so we will prepare the data accordingly


```python
#dateofjourney
df['DateDayNumber']=df['Date_of_Journey'].str.split('/').str[0]
df['Monthnumber']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
#now converting the datatype from object to int
df['DateDayNumber']=df['DateDayNumber'].astype(int)
df['Monthnumber']=df['Monthnumber'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()
```

```python
#route because we have no of stops column
df.drop('Route',axis=1,inplace=True)
```

```python
#arrivaltime and Duration
df['Arrival_hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration'].unique()
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df.loc[df['Duration_hour'] == '5m', 'Duration_hour'] = 0
```

```python
df['Duration_hour'].astype(int)
```

```python
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df
```

```python
df['Duration_min'].fillna(0,inplace=True)
```

```python
for i in ['Arrival_hour','Arrival_min','Duration_hour','Duration_min']:
    df[i]=df[i].astype(int)

df.info()

```

```python
df.head()
```

```python
df.head()
```

```python
#dep time
df['Dep_Time_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_Time_hour']=df['Dep_Time_hour'].astype(int)
df['Dep_Time_min']=df['Dep_Time_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Total stops
df['Total_Stops'].mode()
```

```python
# encode it as a Target label Encoding
df['Total_Stops'].unique()
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4,np.nan:1})
```

```python
#Additional info ,Airlien,Source and Destination
# encode it by OHEncoding
df['Additional_Info'].unique()
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
#merge this data frame with df
df2=pd.DataFrame(encoder.fit_transform(df[['Additional_Info','Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
df2
```

```python
df
```

```python
df3=pd.merge(df,df2,left_index=True, right_index=True)
df3.drop(df[['Airline','Source','Destination','Additional_Info']],axis=1,inplace=True)
```

```python
#final this is the data ready to train model
df3.head()
```
